```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data.head()
```

```python
data.dtypes
```

```python
df_out = data.apply(lambda x: list(x.dropna().values), axis=1).tolist()

Encoder = TransactionEncoder()
dataset = Encoder.fit(df_out).transform(df_out)
dataset = dataset.astype('int')
print(dataset)
```

```python
df = pd.DataFrame(dataset,columns= Encoder.columns_)
df
```

```python
for col in df.columns:
    print(col)
```

```python
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets.sort_values(by=['support'],ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

We can interpret the result of the first rule as:

The support is 0.07 calculated by dividing the number of transactions containing mineral water and chocolate by the total number of transactions.

The confidence level is 0.35 shows that out of all the transactions that contain chocolate, 0.35 transactions contain mineral water too.

The lift 1.19 tells us that mineral water is 1.19 times more likely to be bought by the customer who also buys chocolate.

3. What are some limitations or challenges of Association rules mining?

Limitations and Challenges of Association Rule Mining

Scalability:Mining association rules can be expensive, especially in large datasets, as the number of itemsets increases.

Redundancy:Many generated rules can be redundant or uninteresting. Removing insignificant rules can be challenging.

Threshold Sensitivity:The results of association rule mining heavily depend on the chosen support and confidence thresholds. Setting these thresholds too high may result in missing patterns, where as setting them too low may produce an large number of rules.

Actionability:Not all identified rules are actionable or useful in a practical context.

Dynamic Data:In environments where data is constantly changing, maintaining up-to-date association rules can be challenging and may require frequent recalculations.
