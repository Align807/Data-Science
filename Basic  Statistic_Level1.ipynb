Explain the concept of standardization (z-score normalization).


```python
from scipy.stats import zscore
import numpy as np

```

```python
Volume_zscore=zscore(df['Volume'])
Volume_zscore
```

```python
df['Volume_zscore']=Volume_zscore
df[["Volume","Volume_zscore"]]
```

```python
avg_price_zscore=zscore(df['avg_price'])
avg_price_zscore
```

```python
df['avg_price_zscore']=avg_price_zscore
df[['avg_price','avg_price_zscore']]
```

```python
total_sale_value_zscore=zscore(df['total_sale_value'])
total_sale_value_zscore
```

```python
df['total_sale_value_zscore']=total_sale_value_zscore
df[['total_sale_value','total_sale_value_zscore']]
```

```python
discount_rate_zscore=zscore(df['discount_rate'])
discount_rate_zscore
```

```python
df['discount_rate_zscore']=discount_rate_zscore
df[['discount_rate','discount_rate_zscore']]
```

```python
discount_amount_zscore=zscore(df['discount_amount'])
discount_amount_zscore
```

```python
df['discount_amount_zscore']=discount_amount_zscore
df[['discount_amount','discount_amount_zscore']]
```

```python
net_sales_value_zscore=zscore(df['net_sales_value'])
net_sales_value_zscore
```

```python
df['net_sales_value_zscore']=net_sales_value_zscore
df[['net_sales_value','net_sales_value_zscore']]
```

```python
df =df.drop(['Volume_zscore', 'avg_price_zscore','total_sale_value_zscore','discount_rate_zscore','discount_amount_zscore','net_sales_value_zscore'], axis=1)
```
